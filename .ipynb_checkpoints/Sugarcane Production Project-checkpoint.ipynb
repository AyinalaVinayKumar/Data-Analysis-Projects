```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('List of Countries by Sugarcane Production.csv')
```

```python
df.head()
```

```python
df.shape
```

# Data Cleaning


```python
df['Production (Tons)'] = df['Production (Tons)'].str.replace('.',',')
df['Production per Person (Kg)'] = df['Production per Person (Kg)'].str.replace('.',',')
df['Acreage (Hectare)'] = df['Acreage (Hectare)'].str.replace('.',',')
df['Yield (Kg / Hectare)'] = df['Yield (Kg / Hectare)'].str.replace('.',',')
```

```python
df.head()
```

```python
df.rename(columns = {'Production (Tons)' : 'Production(Tons)'},inplace = True)
df.rename(columns = {'Production per Person (Kg)' : 'Production_per_Person(Kg)'},inplace = True)
df.rename(columns = {'Acreage (Hectare)' : 'Acreage(Hectare)'},inplace = True)
df.rename(columns = {'Yield (Kg / Hectare)' : 'Yield(Kg/Hectare)'},inplace = True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df[df['Acreage(Hectare)'].isnull()]
```

```python
df = df.dropna().reset_index()
```

```python
df.head()
```

```python
df.drop(['index','Unnamed: 0'],axis = 1,inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Ensure the column is a string, remove commas, and convert to float
df["Production(Tons)"] = df["Production(Tons)"].astype(str).str.replace(',', '').astype(float)
df["Production_per_Person(Kg)"] = df["Production_per_Person(Kg)"].astype(str).str.replace(',', '').astype(float)
df["Acreage(Hectare)"] = df["Acreage(Hectare)"].astype(str).str.replace(',', '').astype(float)
df['Yield(kg/Hectare)'] = df['Yield(Kg/Hectare)'].astype(str).str.replace(',','').astype(float)
```

```python
df.drop(['Yield(Kg/Hectare)'],axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.drop(['level_0'],axis = 1, inplace = True)
```

```python
df.nunique()
```

## Checking outliers

```python
plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
sns.boxplot(df['Production(Tons)'])
plt.subplot(2,2,2)
sns.boxplot(df['Production_per_Person(Kg)'])
plt.subplot(2,2,3)
sns.boxplot(df['Acreage(Hectare)'])
plt.subplot(2,2,4)
sns.boxplot(df['Yield(kg/Hectare)'])
```

## Distribution of the columns

```python
plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
sns.distplot(df['Production(Tons)'])
plt.title("Production(Tons)")
plt.subplot(2,2,2)
sns.distplot(df['Production_per_Person(Kg)'])
plt.title("Production_per_person(Kg)")
plt.subplot(2,2,3)
sns.distplot(df['Acreage(Hectare)'])
plt.title("Acreage(Hectare)")
plt.subplot(2,2,4)
sns.distplot(df['Yield(kg/Hectare)'])
plt.title("Yield(Kg/Hectare)")
plt.show()
```

## Which country produces maximum sugarcane?

```python
df_new = df[['Country','Production(Tons)']].set_index('Country')
```

```python
df_new
```

```python
df_new['Production(Tons)_percent'] = df_new['Production(Tons)']*100/df_new['Production(Tons)'].sum()
```

```python
df_new
```

```python
df_new['Production(Tons)_percent'].plot(kind = 'pie',autopct = '%.2f')
```

```python
df_new['Production(Tons)_percent'].head(10).plot(kind = 'bar')
```

```python
ax = sns.barplot(data = df.head(10), x = 'Country',y = 'Production(Tons)')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

## Which country has highest land?

```python
df_acr = df.sort_values('Acreage(Hectare)',ascending = False)
ax = sns.barplot(data = df_acr.head(10), x = 'Country',y = 'Acreage(Hectare)')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

## Which country has highest yield per hectare?

```python
df_yield = df.sort_values('Yield(kg/Hectare)',ascending = False)
ax = sns.barplot(data = df_yield.head(10), x = 'Country',y = 'Yield(kg/Hectare)')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

## Which country has highest production?

```python
df_prod = df.sort_values('Production_per_Person(Kg)',ascending = False)
ax = sns.barplot(data = df_prod, x = 'Country',y = 'Production_per_Person(Kg)')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

## Do number of countries in a Continent effects production of sugarcane?

```python
df_continent['number_of_countries'] = df.groupby('Continent').count()['Country']
```

```python
df_continent
```

```python
continent_names = df_continent.index.to_list()
sns.lineplot(data = df_continent,x = 'number_of_countries',y = 'Production(Tons)')
plt.xticks(df_continent['number_of_countries'],continent_names,rotation = 90)
plt.show()
```
